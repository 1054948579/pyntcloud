# PyntCloud.from_file

The simplest way to read point cloud data is using the PyntCloud's methot `from_file`.

```python
ply_cloud = PyntCloud.from_file("../data/python_bin.ply")
ply_cloud
```

```python
ply_cloud.points
```

```python
ply_cloud.mesh
```

```python
ply_cloud.comments
```

PyntCloud.from_file supports many different point cloud formats and you can find a list along with their principal charateristics in the [** SUPPORTED FILE FORMATS** notebook](supported file formats.ipynb).

In the docs/data folder there are samples of the same point cloud stored in different formats.

```python
pcd_cloud = PyntCloud.from_file("../data/python.pcd")
mat_cloud = PyntCloud.from_file("../data/python.mat")

print(all(pcd_cloud.points == mat_cloud.points))

print(all(ply_cloud.points == mat_cloud.points))
```

# PyntCloud.to_file

The simplest way to write point cloud data is using the PyntCloud's methot `to_file`.

```python
PyntCloud.to_file("out_cloud.npz", points=pcd_cloud.points, important_thing={"Carl": "black", "Lenny": "white"} )
```

```python
writed_cloud = PyntCloud.from_file("out_cloud.npz")
```

```python
print(all(pcd_cloud.points == writed_cloud.points))
```

```python
writed_cloud.important_thing
```
